Task: apply five filters to a photo.

```python
#Packages
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
import numpy as np
```

```python
#Open file
path = 'Birds.jpg'
```

```python
img = plt.imread(path)
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
print(img.shape) #image size (600, 800, 3)
```

```python
#1. Contour
image1 = Image.open(path)
contour_image = image1.filter(ImageFilter.CONTOUR)
contour_image.save("Contour_birds.jpg", "JPEG")
```

```python
#2. EMBOSS filter
image2 = Image.open(path)
EMBOSS_image = image2.filter(ImageFilter.EMBOSS)
EMBOSS_image.save("EMBOSS_birds.jpg", "JPEG")
```

```python
#3. Convert image
image3 = Image.open(path)
Convert_image = image3.convert('L')
Convert_image.save("Convert_birds.jpg", "JPEG")
```

```python
#4. Negative
image4 = plt.imread(path)
image4 = 1 - image4
plt.imshow(image4)
plt.axis('off')
plt.savefig('Negative_birds.png')
plt.close()
```

```python
#5. FIND_EDGES filter
image5 = Image.open(path)
FIND_image = image5.filter(ImageFilter.FIND_EDGES)
FIND_image.save("FIND_birds.jpg", "JPEG")
```
